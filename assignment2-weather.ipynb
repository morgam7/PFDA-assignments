I this assignment I will import a data set and make a plot show two variables

```python
#importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Read in files with pandas
df = pd.read_csv('weatherreadings1.csv', index_col=0) #the first row is not data
df
```

```python

#getting my datasets for x and y axis. Had to look at last semesters work to remember how to do this!

x = df["reportStartDateTime"]
y = df["dryBulbTemperature_Celsius"]


fig, ax = plt.subplots()
ax.set(xlabel="Date and Time", ylabel="Dry Bulb Temperature", title="Temperature over Time")

ax.plot(x,y)

plt.show()
```
