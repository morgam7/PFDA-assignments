In this assignment, I will create a visually appealing plot using a weather dataset. The goal is to plot the variable "dryBulbTemperature_Celsius" against "reportStartDateTime" and demonstrate good practices in data visualization, including formatting and clarity.

```python
#importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Read in files with pandas
df = pd.read_csv('weatherreadings1.csv') 
df
```

# Data Preparation and Plotting
- First, I converted the 'reportStartDateTime' column to datetime format to handle the time-series data correctly. This solved the issue of the dates being bunched up on the x-axis and allowed for better spacing and formatting.
- Next, I selected the variables for the x and y axes. I revisited last semester's work to recall how to structure the datasets for plotting.
- Adjustments such as gridlines, labels, and layout tweaks were added to ensure the plot is clear and visually appealing.

```python
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])

x = df["reportStartDateTime"]
y = df["dryBulbTemperature_Celsius"]


fig, ax = plt.subplots(figsize=(10, 6)) 
ax.plot(x, y, label="Temperature", color="blue", linestyle="-", linewidth=1.5)
ax.set(xlabel="Date and Time", ylabel="Dry Bulb Temperature (Â°C)", title="Temperature Over Time")
ax.grid(True, linestyle="--", alpha=0.6)

plt.xticks(rotation=45)

ax.legend()

plt.tight_layout()
plt.plot(x,y)
plt.show()
```
